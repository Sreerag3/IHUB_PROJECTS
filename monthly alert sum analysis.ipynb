```python
import pandas as pd

data = pd.read_csv('/content/february.csv',error_bad_lines=False)
print(data.shape)
data.head()
```

```python
data.columns
```

```python
data = data[['VehicleNum','me_fcw_count', 'me_hmw_count', 'me_lldw_count', 'me_rldw_count',
             'me_pcw_count']]
data.head()
```

```python
count_lists = ['me_fcw_count','me_hmw_count', 'me_lldw_count',	'me_rldw_count',	'me_pcw_count']
data_ = data.groupby(['VehicleNum'])[count_lists].sum().reset_index()
data_.head(10)
```

```python
sum_row = data_.sum()
data_ = data_.append(sum_row, ignore_index=True)
data_.tail()
```

```python
data_.to_excel("Sum of alerts.xlsx")
```
