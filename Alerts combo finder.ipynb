```python
import numpy as np
import pandas as pd
# from openpyxl import load_workbook
df = pd.read_csv(R"/content/Dev sir headway gap.csv")
# , sheet_name = '2Buses_RHS'
df
```

```python
def get_seconds(time_str):
    # print('Time in hh:mm:ss:', time_str)
    # split in hh, mm, ss
    hh, mm, ss = time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)
# Applying the get_seconds function to time column in dataframe
seconds = df.time.apply(get_seconds)
# Creating a new column in dataframe by name seconds which contains time in seconds format
df["seconds"] = seconds
df
```

```python
# Create two empty rows as a DataFrame
empty_rows = pd.DataFrame(index=[0, 1], columns=df.columns)

# Concatenate the empty rows and the original DataFrame
df = pd.concat([empty_rows, df], ignore_index=True)

df
```
