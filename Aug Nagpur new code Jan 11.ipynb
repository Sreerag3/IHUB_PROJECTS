```python
import pandas as pd
from datetime import datetime
data = pd.read_csv('/content/trip raw report august 23.csv')
df = data.copy()
```

```python
count_lists = ['distance', 'CAS', 'me_fcw_count', 'me_hmw_count', 'me_pcw_count', 'me_lldw_count', 'me_rldw_count']
grouped_data = data.groupby(['groupName', 'VehicleNum', 'startDate', 'driverName'])[count_lists].sum().reset_index()
```

```python
def custom_date_parser(date_str):
    try:
        return datetime.strptime(date_str, '%d-%m-%Y')
    except ValueError:
        return datetime.strptime(date_str,'%Y-%m-%d')

grouped_data['startDate'] = grouped_data['startDate'].apply(custom_date_parser)
dfs = []

day_list = [(1, 8), (9, 16), (17, 24), (25, 31)]

for day in day_list:
    start_date = datetime(2023, 8, day[0])
    end_date = datetime(2023, 8, day[1])
    mask = (grouped_data['startDate'] >= start_date) & (grouped_data['startDate'] <= end_date)
    filtered_df = grouped_data[mask]
    filtered_df.reset_index(drop=True, inplace=True)
    dfs.append(filtered_df)
```

```python
dfs[0].to_excel('Aug cas analysis week 1.xlsx', index=False)
```

```python
import pandas as pd

# List of Excel file paths
excel_file_paths = [
    r'/content/Aug cas main analysis week 1.xlsx',
    r'/content/Aug cas main analysis week 2.xlsx',
    r'/content/Aug cas main analysis week 3.xlsx',
    r'/content/Aug cas main analysis week 4.xlsx',
    # Add more file paths as needed
]

# Read all Excel files into a list of pandas DataFrames
dfs = [pd.read_excel(file_path) for file_path in excel_file_paths]

# Perform the join operation by vertically concatenating the DataFrames
merged_df = pd.concat(dfs, axis=0, ignore_index=True)

# Save the result to a new Excel file
merged_df.to_excel('combined file.xlsx', index=False)
```

```python
count_lists = ['distance', 'CAS', 'me_fcw_count', 'me_hmw_count', 'me_pcw_count', 'me_lldw_count', 'me_rldw_count']
finaldata_ = merged_df.groupby(['groupName','driverName'])[count_lists].sum().reset_index()
print(finaldata_.shape)
finaldata_.head()
```
