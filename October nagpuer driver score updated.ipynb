```python
import pandas as pd
from datetime import datetime

data = pd.read_csv('/content/Trip Raw Report Oct.csv')
df = data.copy()

def custom_date_parser(date_str):
    try:
        return datetime.strptime(date_str, '%m/%d/%Y')
    except ValueError:
        return datetime.strptime(date_str, '%m-%d-%Y')

df['endDate'] = df['endDate'].apply(custom_date_parser)
dfs = []

day_list = [(1, 8), (9, 16), (17, 24), (25, 31)]

for day in day_list:
    start_date = datetime(2023, 10, day[0])
    end_date = datetime(2023, 10, day[1])
    mask = (df['endDate'] >= start_date) & (df['endDate'] <= end_date)
    filtered_df = df[mask]
    filtered_df.reset_index(drop=True, inplace=True)
    dfs.append(filtered_df)

dfs[0].to_excel('Oct 1 - 8.xlsx')

# Read the Excel file containing the list of vehicle numbers to drop
vehicle_numbers_to_drop = pd.read_excel('/content/week 1.xlsx')['VehicleNum'].tolist()

# Remove rows with specified vehicle numbers from the first weekly DataFrame
dfs[0] = dfs[0][~dfs[0]['VehicleNum'].isin(vehicle_numbers_to_drop)]

print(dfs[0].shape)
```

```python
combined_df = pd.concat([dfs[0],dfs[1], dfs[2], dfs[3]], axis=0, ignore_index=True)
combined_df.head()
```
