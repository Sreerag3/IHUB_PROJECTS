```python
import pandas as pd

data = pd.read_csv('/content/Trip_raw_report_september[1].csv')
print(data.shape)
data.head()
```

```python
data = data[['endDate', 'VehicleNum', 'groupName', 'driverName', 'distance', 'me_fcw_count', 'me_hmw_count',
             'me_pcw_count', 'me_lldw_count', 'me_rldw_count']]
data.head()
```

```python
df = data.copy()
print(df.shape)
```

```python
data['endDate'].value_counts()
```

```python
from datetime import datetime

def custom_date_parser(date_str):

    return datetime.strptime(date_str, '%Y-%m-%d')

df['endDate'] = df['endDate'].apply(custom_date_parser)
dfs = []

day_list = [(1,22),(23,30)]

for day in day_list:

  start_date = datetime(2023, 9, day[0])
  end_date = datetime(2023, 9, day[1])
  # print(start_date)
  mask = (df['endDate'] >= start_date) & (df['endDate'] <= end_date)
  filtered_df = df[mask]

  filtered_df.reset_index(drop=True, inplace=True)
  dfs.append(filtered_df)
```
