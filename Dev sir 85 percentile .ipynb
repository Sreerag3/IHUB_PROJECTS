```python
import pandas as pd
import numpy as np

# Read the data from the CSV file
data = pd.read_csv('/content/New SI values.csv')

# Select relevant columns
data = data[['FID', 'New SI']]

# Aggregate data without uniquing
aggregated_data = data.groupby('FID').agg({'New SI': lambda x: ','.join(map(str, x))}).reset_index()

# Calculate and print the 85th percentile for each row
for index, row in aggregated_data.iterrows():
    fid = row['FID']
    new_si_values = list(map(float, row['New SI'].split(',')))
    print(new_si_values)

    if len(new_si_values) > 0:
        # Calculate the 85th percentile
        percentile_85 = np.percentile(new_si_values, 85)

        print(f"FID: {fid}")
        print(f"85th Percentile of 'New SI': {percentile_85}")
        print("===")
        aggregated_data['Agg'] = percentile_85
    else:
        print(f"Not enough data for FID: {fid}")

# Save the result to an Excel file
aggregated_data.to_excel("aggregated_with_percentile.xlsx", index=False)
```
