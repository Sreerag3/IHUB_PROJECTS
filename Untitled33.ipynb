```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('/content/Dev sir headway gap.csv')

# Sort the DataFrame based on the specified columns in ascending order
data = data.sort_values(by=['vehicleNum', 'date', 'time'], axis=0)

# Print the shape and the first few rows of the sorted DataFrame
print(data.shape)
print(data.head())

# Select specific columns after sorting
data = data[['vehicleNum', 'date', 'time', 'alertType']]
data.head()

# Calculate the time difference between consecutive rows
data['time_difference'] = pd.to_datetime(data['time']).diff().dt.total_seconds()

# Print the previous row along with rows that meet the criteria
for index, row in data.iterrows():
    if row['time_difference'] <= 5 or pd.isnull(row['time_difference']):
        # Print the previous row
        if index > 0:
            print(data.iloc[index - 1])

        # Print the current row
        print(row)

# Extract rows where the time difference is 5 seconds or less
filtered_data = data[(data['time_difference'] <= 5) | (data['time_difference'].isnull())]
print(filtered_data.shape)

# Save the filtered data to Excel
filtered_data.to_excel('filtered_data.xlsx')
```

```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('/content/Dev sir headway gap.csv')

# Sort the DataFrame based on the specified columns in ascending order
data = data.sort_values(by=['vehicleNum', 'date', 'time'], axis=0)

# Select specific columns after sorting
data = data[['vehicleNum', 'date', 'time', 'alertType']]

# Calculate the time difference between consecutive rows
data['time_difference'] = pd.to_datetime(data['time']).diff().dt.total_seconds()

# Print every previous row for each row that meets the criteria
for index, row in data.iterrows():
    if row['time_difference'] <= 5 or pd.isnull(row['time_difference']):
        # Print all previous rows only once
        print(data.loc[:index - 1])

        # Print the current row
        print(row)

# Extract rows where the time difference is 5 seconds or less
filtered_data = data[(data['time_difference'] <= 5) | (data['time_difference'].isnull())]

# Save the filtered data to Excel
filtered_data.to_excel('filtered_data.xlsx')
```
