```python
import pandas as pd
from datetime import datetime

data = pd.read_csv('/content/Trip Raw Report_November.csv')
df = data.copy()

def custom_date_parser(date_str):
    try:
        return datetime.strptime(date_str, '%d-%m-%Y')
    except ValueError:
        return datetime.strptime(date_str,'%Y-%m-%d')

df['endDate'] = df['endDate'].apply(custom_date_parser)
dfs = []

day_list = [(1, 8), (9, 16), (17, 24), (25, 30)]

for day in day_list:
    start_date = datetime(2023, 11, day[0])
    end_date = datetime(2023, 11, day[1])
    mask = (df['endDate'] >= start_date) & (df['endDate'] <= end_date)
    filtered_df = df[mask]
    filtered_df.reset_index(drop=True, inplace=True)
    dfs.append(filtered_df)
```

```python
count_lists = ['distance', 'CAS', 'me_fcw_count', 'me_hmw_count', 'me_pcw_count', 'me_lldw_count', 'me_rldw_count']

data1_ = dfs[0].groupby(['groupName', 'VehicleNum','driverName'])[count_lists].sum().reset_index()

# Read the Excel file containing the list of faulty vehicle numbers
faulty_vehicle_list = pd.read_excel('/content/week 1 nov faulty.xlsx')['VehicleNum'].tolist()

# Remove rows only when 'CAS' is 0 and 'VehicleNum' is in the faulty vehicle list
data1_ = data1_[~((data1_['CAS'] == 0) & (data1_['VehicleNum'].isin(faulty_vehicle_list)))]

print(data1_.shape)
```

```python
count_lists = ['distance', 'CAS', 'me_fcw_count', 'me_hmw_count', 'me_pcw_count', 'me_lldw_count', 'me_rldw_count']

data2_ = dfs[1].groupby(['groupName','VehicleNum', 'driverName'])[count_lists].sum().reset_index()

# Read the Excel file containing the list of faulty vehicle numbers
faulty_vehicle_list = pd.read_excel('/content/week 2 nov faulty.xlsx')['VehicleNum'].tolist()

# Remove rows only when 'CAS' is 0 and 'VehicleNum' is in the faulty vehicle list
data2_ = data2_[~((data1_['CAS'] == 0) & (data2_['VehicleNum'].isin(faulty_vehicle_list)))]

print(data2_.shape)
```

```python
count_lists = ['distance', 'CAS', 'me_fcw_count', 'me_hmw_count', 'me_pcw_count', 'me_lldw_count', 'me_rldw_count']

data3_ = dfs[2].groupby(['groupName','VehicleNum', 'driverName'])[count_lists].sum().reset_index()

# Read the Excel file containing the list of faulty vehicle numbers
faulty_vehicle_list = pd.read_excel('/content/week 3 nov faulty.xlsx')['VehicleNum'].tolist()

# Remove rows only when 'CAS' is 0 and 'VehicleNum' is in the faulty vehicle list
data3_ = data3_[~((data3_['CAS'] == 0) & (data3_['VehicleNum'].isin(faulty_vehicle_list)))]

print(data3_.shape)
```

```python
count_lists = ['distance', 'CAS', 'me_fcw_count', 'me_hmw_count', 'me_pcw_count', 'me_lldw_count', 'me_rldw_count']

data4_ = dfs[3].groupby(['groupName','VehicleNum', 'driverName'])[count_lists].sum().reset_index()

# Read the Excel file containing the list of faulty vehicle numbers
faulty_vehicle_list = pd.read_excel('/content/week 4 nov faulty.xlsx')['VehicleNum'].tolist()

# Remove rows only when 'CAS' is 0 and 'VehicleNum' is in the faulty vehicle list
data4_ = data4_[~((data4_['CAS'] == 0) & (data4_['VehicleNum'].isin(faulty_vehicle_list)))]

print(data4_.shape)
```

```python
combined_df = pd.concat([data1_,data2_,data3_,data4_], axis=0, ignore_index=True)
combined_df.head()
```

```python
count_lists = ['distance', 'CAS', 'me_fcw_count', 'me_hmw_count', 'me_pcw_count', 'me_lldw_count', 'me_rldw_count']
finaldata_ = combined_df.groupby(['groupName','driverName'])[count_lists].sum().reset_index()
print(finaldata_.shape)
finaldata_.head()
```

```python
name_list = finaldata_['groupName'].unique()
print(name_list)

for g_value in name_list:

  g_df = finaldata_[finaldata_['groupName'] == g_value]
  f_name = f"group_{g_value}.xlsx"
  g_df.to_excel(f_name, index=False)
```
