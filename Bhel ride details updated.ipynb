```python
import pandas as pd

data = pd.read_csv('/content/BHEL Adas & Non adas buses analysis.csv')
print(data.shape)
data.head()
```

```python
data.columns
```

```python
data = data[['SERVICE No.','OPTD_ KMS','VEHICLE No.','ADAS bus','Non ADAS bus','Corridor']]
data.head()
```

```python
df = data.copy()
print(df.shape)
```

```python
count_lists = ['OPTD_ KMS','ADAS bus','Non ADAS bus']
data_ = df.groupby(['SERVICE No.','Corridor'])[count_lists].sum().reset_index()
print(data_.shape)
```

```python
vehicle_numbers = df['VEHICLE No.']

    # Use Counter to count occurrences of each unique vehicle
vehicle_counter = Counter(vehicle_numbers)

    # Display the result
print(f"Unique vehicles and their counts in column '{'VEHICLE No.',}':")
for vehicle, count in vehicle_counter.items():
print (f"{vehicle}: {count} vehicles")
```
